```python
import pandas as pd
```

```python
df = pd.read_csv("log.csv")
df.sort_values(by='N')
```

```python
df_800 = df[df.N==800]
df_1600 = df[df.N==1600]
df_2400 = df[df.N==2400]
list_df = [df_800, df_1600, df_2400]
for d_f in list_df:
    print(d_f)
```

```python
for i, d_f in enumerate(list_df):
    ax = d_f.plot.bar(x='nbl', y='time', rot=0, title=f'N={(i+1)*800}')
```
